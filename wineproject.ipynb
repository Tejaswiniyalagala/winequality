```python
dataset=pd.read_csv('wine.csv')
```

```python
x=dataset.iloc[:,0:11].values
```

```python
from sklearn.model_selection import train_test_split                
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test
```

```python
import keras 
from keras.models import Sequential 
from keras.layers import Dense 
```

```python
model=Sequential()
```

```python
x.shape
```

```python
model.add(Dense(input_dim=11,init="random_uniform",activation="relu",output_dim=8)) 
```

```python
model.add(Dense(output_dim=8,init="random_uniform",activation="relu"))
```

```python
model.add(Dense(output_dim=1,init="random_uniform"))
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
model.fit(x_train,y_train,epochs=400,batch_size=10)
```

```python
model.predict(x_test)
```
